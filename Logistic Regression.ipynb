# Model Building

```python
x = pd.concat([bank1.iloc[:,0:11], bank1.iloc[:,12:]], axis=1)
y = bank1.iloc[:,11]
```

```python
classifier = LogisticRegression()
classifier.fit(x,y)
```

# Model Predictions

```python
# Predict for x dataset
y_pred = classifier.predict(x)
y_pred
```

```python
y_pred_df = pd.DataFrame({'actual_y':y, 'y_pred_prob':y_pred})
y_pred_df
```

```python
# with the code below, we are finding out the cases where the model predictions match the actual values in the data, and cross-checking the accuracy of the model
```

```python
len(y_pred_df[y_pred_df['actual_y']==y_pred_df['y_pred_prob']])/len(y_pred_df)
```

# Testing Model Accuarcy

```python
# Confusion Matrix for the model accuracy
confusion_matrix = confusion_matrix(y,y_pred)
confusion_matrix
```

```python
#left diagonal talks about correct prediction by model.
#right diagonal talks about incorrectly classified observations or errors.
#the lesser the values in the right diagonal, the better is your model.
```

```python
# model accuracy (manually)
(39152+1238)/(39152+770+4051+1238)
```

```python
TP = len(y_pred_df[(y_pred_df['actual_y']==1) & (y_pred_df['y_pred_prob']==1)])
TN = len(y_pred_df[(y_pred_df['actual_y']==0) & (y_pred_df['y_pred_prob']==0)])
FP = len(y_pred_df[(y_pred_df['actual_y']==0) & (y_pred_df['y_pred_prob']==1)])
FN = len(y_pred_df[(y_pred_df['actual_y']==1) & (y_pred_df['y_pred_prob']==0)])

print('True Positives : ',TP)
print('True Negatives : ',TN)
print('False Positives : ',FP)
print('False Negatives : ',FN)
```

```python
#Checking for sensitivity/Recall
Sensitivity = (TP/(TP+FN))
print('Sensitivity ',Sensitivity)
```

```python
#Checking for Specificity
Specificity = (TN/(TN+FP))
print('Specificity ',Specificity)
```

```python
#Checking for Precision
Precision = (TP/(TP+FP))
print('Precision ', round(Precision,4))
```

```python
#Checking for F-score
F_Score  = (2*Precision*Sensitivity)/(Precision+Sensitivity)
print('F_Score ',round(F_Score,4))
```

```python
#Classification report
print(classification_report(y,y_pred))
```

```python
# As accuracy = 0.89336, which is greater than 0.5; thus [:,1] Threshold value>0.5=1 else [:,0] Threshold value<0.5=0
classifier.predict_proba(x)[:,1]
```

```python
# Roc curve plotting and finding AUC value
fpr, tpr, thresholds = roc_curve(y, classifier.predict_proba(x)[:,1])

auc = roc_auc_score(y, y_pred)

plt.plot(fpr, tpr, color='red', label = 'logit model(area = %0.2f)'%auc)
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
plt.show()
print('\n auc accuracy:', auc)
```
