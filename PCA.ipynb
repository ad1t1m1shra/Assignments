Perform Principal component analysis and perform clustering using first 3 principal component scores (both heirarchial and k mean clustering(scree plot or elbow curve) and obtain 
optimum number of clusters and check whether we have obtained same number of clusters with the original data 
(class column we have ignored at the begining who shows it has 3 clusters)df

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
import warnings
warnings.filterwarnings('ignore')
```

```python
wine = pd.read_csv('wine.csv')
wine
```

```python
wine.describe()
```

```python
wine['Type'].value_counts()
```

```python
wine2 = wine.iloc[:,1:]
wine2
```

```python
wine2.shape
```

```python
wine2.info()
```

```python
wine_ary = wine2.values
wine_ary
```

```python
wine_norm = scale(wine_ary)
wine_norm
```

# PCA Implementation

```python
pca = PCA(n_components=13)
wine_pca = pca.fit_transform(wine_norm)
wine_pca
```

```python
pca.components_
```

```python
var = pca.explained_variance_ratio_
var
```

```python
var1 = np.cumsum(np.round(var, 4)*100)
var1
```

```python
plt.plot(var1, color='green')
```

```python
PC_values = np.arange(pca.n_components_) + 1
plt.plot(PC_values, pca.explained_variance_ratio_, 'o-', linewidth=2, color='green')
plt.title('Scree PLot')
plt.xlabel('Principle Components')
plt.ylabel('Variance Explained')
plt.show()
```

```python
plt.bar(range(1, len(var)+1),var)
plt.xlabel('Number of components')
plt.ylabel('Variance (%)')
plt.title('Explained variance by each component')
```

```python
finaldf = pd.concat([wine['Type'], pd.DataFrame(wine_pca[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
finaldf
```

```python
fig = plt.figure(figsize=(16,10))
sns.scatterplot(data=finaldf)
```

# Checking with other Clustering Algorithms
# 1. Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
plt.figure(figsize=(20,8))
dendro = sch.dendrogram(sch.linkage(wine_norm, 'complete'))
```

```python
hcluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='complete')
hcluster
```

```python
y = pd.DataFrame(hcluster.fit_predict(wine_norm), columns=['cluster_hid'])
y['cluster_hid'].value_counts()
```

```python
winehc = wine.copy()
winehc['cluster_hid'] = hcluster.labels_
winehc
```

```python
winehc['cluster_hid'].value_counts()
```

```python
winehc.groupby('cluster_hid').agg(['mean'])
```

```python
display('After Applying PCA and Hierarchical CLustering on Dataset',winehc.groupby('cluster_hid').agg(['mean']),
        'Original Classified Dataset',wine.groupby('Type').agg(['mean']))
```

# 2. K-Means Clustering

```python
from sklearn.cluster import KMeans
```

```python
# As we already have normalized data
# Use Elbow Graph to find optimum number of  clusters (K value) from K values range
# The K-means algorithm aims to choose centroids that minimise the inertia, or within-cluster sum-of-squares criterion WCSS 
# random state can be anything from 0 to 42, but the same number to be used everytime,s
```

```python
#within cluster sum of squares criterion
```

```python
wcss = []
for i in range(1, 6):
    kmeans = KMeans(n_clusters=i, random_state=2)
    kmeans.fit(wine_norm)
    wcss.append(kmeans.inertia_)
```

Plot K values range vs wcss to get elbow graph for choosing k(no. of clusters)

```python
plt.plot(range(1,6), wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

# Build cluster algorithm using K = 3

```python
clusterK3 = KMeans(3, random_state=30).fit(wine_norm)
clusterK3
```

```python
clusterK3.labels_
```

Assign Clusters to the dataset

```python
wineKm = wine.copy()
wineKm['cluster_kid'] = clusterK3.labels_
wineKm
```

```python
wineKm['cluster_kid'].value_counts()
```

```python
wineKm.groupby('cluster_kid').agg(['mean'])
```

```python
display('After Applying PCA and Kmens CLustering on Dataset',wineKm.groupby('cluster_kid').agg(['mean']),
        'Original Classified Dataset',wine.groupby('Type').agg(['mean']))
```
